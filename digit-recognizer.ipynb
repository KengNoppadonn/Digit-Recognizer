```python
import pandas as pd
import numpy as np
data=pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
X=data.drop(columns='label')
Y=data['label']
test_data=pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
test_data
```

```python
import matplotlib.pyplot as plt, matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
i=890
img=X.iloc[i].values
img=img.reshape((28,28))
plt.imshow(img)
plt.title(Y[i])
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)                    ## define the nearest centroid model
model.fit(X, Y)                               ## train the model with the dataset
```

```python
label  = model.predict(test_data)            ## predict a label for the example
```

```python
df=test_data
df=pd.DataFrame({'Label':label})
df['ImageId'] = np.arange(len(df))+1
df
```

```python
df.to_csv('submission.csv', index=False)
```

```python
import pickle

with open('Digit_Recognizer.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
output =pd.DataFrame()
output.to_csv('submission.csv', index=False)
```
